```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
stock_data = yf.download('spy', start="2022-07-05", end="2023-07-05")
stock_data1 = yf.download('nvda', start = '2022-07-05', end= '2023-07-05')
```

```python
df = pd.DataFrame(stock_data['Close'])
df1 = pd.DataFrame(stock_data1['Close'])
```

```python
df['1D returns'] = df['Close'].pct_change()
df1['1D returns'] = df1['Close'].pct_change()
```

```python
df.dropna(inplace = True)
df1.dropna(inplace=True)
```

```python
spy = df
stock = df1
```

```python
x = spy['1D returns']
y = stock['1D returns']
```

```python
plt.scatter(spy['1D returns'],stock['1D returns'])
```

```python
correlation = x.corr(y)
```

```python
Beta = correlation * (y.std()/x.std())
```
